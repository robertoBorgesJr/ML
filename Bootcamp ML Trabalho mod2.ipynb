```
df_bike_sharing['season'] = df_bike_sharing['season'].astype('category')
```

```
df_bike_sharing['season'].unique()
```

```
df_seasons = df_bike_sharing[['season', 'total_count']]
```

```
import matplotlib.pyplot as plt 
```

```
import seaborn as sns #biblioteca utilizada para criar gráficos mais "bonitos"
```

```
sns.countplot(x='season', data=df_seasons)
plt.title('Teste');
```

```
df_bike_sharing.hist('hour');  #seleciona a coluna de total de gastos para realizar o histograma
                                 # os "bins" indicam a quantidade de grupos que se deseja dividir os dados
plt.title('Distribuição dos clientes pelo gasto anual em produtos derivados do leite');# adiciona o título ao gráfico (histograma)
plt.xlabel('Gasto total em produtos derivados do leite');
```

```
df_bike_sharing.hist('weekday');  #seleciona a coluna de total de gastos para realizar o histograma
                                 # os "bins" indicam a quantidade de grupos que se deseja dividir os dados
plt.title('Distribuição dos clientes pelo gasto anual em produtos derivados do leite');# adiciona o título ao gráfico (histograma)
plt.xlabel('Gasto total em produtos derivados do leite');
```
