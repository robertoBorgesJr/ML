```
labels = np.array(df['Class'])
```

```
# salva a ordem das features antes de dropar a coluna Class
feature_list = list(df.columns)
```

```
df = df.drop('Class', axis = 1)
```

```
df.columns
```

```
# transforma em array
data = np.array(df)
```

```
# separa os dados em treino e teste
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size = 0.25, random_state = 42)
```

```
# aplica o SVM
from sklearn.svm import SVC

classifier = SVC(kernel='rbf')

classifier.fit(train_data, train_labels)
```

```
predictions_label = classifier.predict(test_data)
```

```
from sklearn import metrics
```

```
metrics.classification_report(test_labels, predictions_label)
```

```
print('Acurácia: ', metrics.accuracy_score(test_labels, predictions_label))
print('Precision: ', metrics.precision_score(test_labels, predictions_label))
print('Recall: ', metrics.recall_score(test_labels, predictions_label))
print('F1: ', metrics.f1_score(test_labels, predictions_label))
print('AUROC: ', metrics.roc_auc_score(test_labels, predictions_label))
```

```
# criando baseline para comparação
baseline_preds = np.random.choice([0,1], size = len(test_labels))
```

```
print('Acurácia: ', metrics.accuracy_score(test_labels, baseline_preds))
#print('Precision: ', metrics.precision_score(test_labels, baseline_preds))
#print('Recall: ', metrics.recall_score(test_labels, baseline_preds))
#print('F1: ', metrics.f1_score(test_labels, baseline_preds))
print('AUROC: ', metrics.roc_auc_score(test_labels, baseline_preds))
```
